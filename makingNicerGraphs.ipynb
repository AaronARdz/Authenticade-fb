```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3,101)

# print(x)

plt.plot(x,x,label='y=x')
plt.plot(x,x**2,label='y=x**2')
plt.plot(x,x**3, label='y=x**3')

plt.xlabel('x')
plt.ylabel('y=f(x)')

# plt.xlim([-3,3])
plt.xlim(x[0],x[-1])
plt.ylim([-10,10])

plt.title('nicer graph')

# arreglar los bordes cuadrados de la grafica
plt.gca().set_aspect(1/plt.gca().get_data_ratio())
print(plt.gca().get_data_ratio())

plt.legend()
# or plt.legend(['x','x**2','x**3'])
plt.show()
```

```
fig,ax = plt.subplots()

print(ax)

ax.plot(x,x,label='y=x')
ax.plot(x,x**2,label='y=x**2',color=[0.5,.2,0.5])
ax.plot(x,x**3, label='y=x**3',color=[0.5,1,0.5])
ax.legend()

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('hi')

ax.set_xlim([-3,3])
ax.set_aspect(1/ax.get_data_ratio())

ax.grid()
plt.show()
```
